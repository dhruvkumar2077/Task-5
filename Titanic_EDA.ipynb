```python
#Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load Dataset from URL (Titanic train.csv)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
#Basic Info
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
print(df.info())
```

```python
print("\nFirst 5 Rows:")
print(df.head())
```

```python
#Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
#Statistical Summary
print("\nStatistical Summary:")
print(df.describe())
```

```python
#Univariate Analysis
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
#Bivariate Analysis
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Age vs Passenger Class")
plt.show()
```

```python
#Multivariate Analysis
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Key Insights
print("\n📌 Key Insights:")
print("1. Women had a higher survival rate than men.")
print("2. Passengers in 1st class had the highest survival rate.")
print("3. Fare has a positive correlation with survival.")
print("4. Younger passengers were more common in 3rd class.")
```
